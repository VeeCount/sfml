```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

def limit2char( limit_bal ):
    if limit_bal <= 10000:
        return 'A'
    elif limit_bal <=100000:
        return 'B'
    elif limit_bal <=200000:
        return 'C'
    elif limit_bal <=400000:
        return 'D'
    elif limit_bal <=700000:
        return 'E'
    else:
        return 'F'

df['LIMIT_CHAR_ID'] = df['LIMIT_BAL'].map(limit2char)
df[['ID', 'LIMIT_BAL', 'LIMIT_CHAR_ID']].head(10)
```

```python
# (8) построить среднее значение кредитного лимита для каждого вида образования 
# и для каждого пола
# график необходимо сделать очень широким (на весь экран)
import matplotlib.pyplot as plt

# Вариант 1
fig = plt.figure(figsize=(20,4))
ax = plt.subplot(1,2,1)
df.groupby('SEX').LIMIT_BAL.mean().plot(ax=ax)
ax = plt.subplot(1,2,2)
df.groupby('EDUCATION').LIMIT_BAL.mean().plot(ax=ax)
```

```python
# (9) построить зависимость кредитного лимита и образования только для одного из полов
# Вариант 1
fig = plt.figure(figsize=(20,10))
sex_dict = {1: 'Male', 2: 'Female'}

for i, t in sex_dict.items():
    ax = plt.subplot(1,2,i)
    ax.set_title(t)
    df[ df.SEX == i ].groupby('EDUCATION').LIMIT_BAL.mean().plot()
```

```python
# Вариант 2
df['LIMIT_BAL'][ df['SEX'] == 1 ].plot.hist(alpha=0.5, label="Male")
df['LIMIT_BAL'][ df['SEX'] == 2 ].plot.hist(alpha=0.5, label="Female")
plt.legend()
```
